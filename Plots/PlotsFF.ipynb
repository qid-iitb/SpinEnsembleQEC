```python
d000 = np.loadtxt('Data00_LocalEncoded_20_10_2000.txt', delimiter='\t', unpack = True)
#d010 = np.loadtxt('Data01_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
#d100 = np.loadtxt('Data100_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
#d101 = np.loadtxt('Data101_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
d110 = np.loadtxt('Data110_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
d111 = np.loadtxt('Data111_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
d200 = np.loadtxt('Data20_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
#d210 = np.loadtxt('Data21_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
dun0 = np.loadtxt('DataUn_LocalEncoded_20_10_2000.txt', delimiter='\t', unpack = True)
```

```python
t = d000[0][:-1]
d00 = d000[1][:-1]#[(d000[1][i] + d010[1][i])/2 for i in range(len(t))]
#d10 = [(d100[3][i] + d101[3][i])/2 for i in range(len(t))]
d11 = [(d110[4][i] + d111[4][i])/2 for i in range(len(t))]
d20 = d200[6][:-1]#[(d200[6][i] + d210[6][i])/2 for i in range(len(t))]
dun = dun0[1][:-1]
```

```python
fig, ax = plt.subplots(figsize=(6,5.25))

ax.plot(t,dun,marker='D',markevery=10,color='red',linestyle='dotted',label=r'Unencoded')

ax.plot(t,d00,marker='v',markevery=10,color='blue',linestyle='dashed',label=r'Encoded')

#ax.plot(t,d10,marker='^',markevery=10,color='red',linestyle='solid',label='Encoded w/error \& QEC')

ax.plot(t,d11,marker='^',markevery=10,color='black',linestyle='solid',label=r'\noindent Encoded w/QEC')

ax.plot(t,d20,marker='.',markevery=10,color='green',linestyle='solid',label=r'\noindent Encoded w/QEC \\and Extraction')

#axins = ax.inset_axes(
#    [0.6, 0.6, 0.3, 0.3],
#    xlim=(-0.01, 0.05), ylim=(0.9999, 1.0001), yticklabels=[' ',' ', 1])
#
#axins.plot(t,dun,marker='D',markevery=1,color='blue',linestyle='dotted')
#
#axins.plot(t,d00,marker='v',markevery=1,color='red',linestyle='dashed')
#
##axins.plot(t,d10,marker='^',markevery=1,color='red',linestyle='solid',label='Encoded w/error \& QEC')
#
#axins.plot(t,d11,marker='^',markevery=1,color='green',linestyle='solid')
#
#axins.plot(t,d20,marker='.',markevery=1,color='green',linestyle='solid')
#
#ax.indicate_inset_zoom(axins, edgecolor="black")

ax.set_xlim((0,1.0))
ax.set_ylim((0,1.04))
#ax.set_ylim((0.999,1.0001))
#ax.set_yticks([0,1e-2,2e-2,3e-2,4e-2])
#ax.set_xlim((0,1e-1))
ax.set_xlabel(r'$\gamma_{m}t$',fontsize=20,labelpad=5)
ax.set_ylabel(r'$\mathcal{F}$',fontsize=20,labelpad=20)

#ax.text(0.7, 0.7, r"$\gamma_0 = \gamma_{\pm1} = 1$", fontsize = 14)
#plt.legend(loc='right', bbox_to_anchor=(1.65,1/2),frameon=False,fontsize=17)
plt.legend(fontsize = 13, loc='right', bbox_to_anchor=(1,0.7),frameon=False)
plt.ticklabel_format(axis='y',style='sci',scilimits=(0,0))
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)


#plt.title(r'$N = {}, J = {}, \Delta ~T = 0.01, ~ \gamma_{{\downarrow}}$ = {}'.format(N,J,dr))
plt.savefig("CollectiveQEC.pdf",bbox_inches='tight',dpi=600)
plt.show()
```
