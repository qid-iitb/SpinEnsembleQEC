```python
import scipy as scipy
import numpy as np
import matplotlib.pyplot as plt
import pylab
from matplotlib import rc
#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
plt.rc('font',**{'family':'serif','sans-serif':['Computer Modern Sans serif'],'size':16}) # The default font in LaTex
# font.family        : serif
# font.serif         : Times, Palatino, New Century Schoolbook, Bookman, Computer Modern Roman
# font.sans-serif    : Helvetica, Avant Garde, Computer Modern Sans serif
# font.cursive       : Zapf Chancery
# font.monospace     : Courier, Computer Modern Typewriter
# text.usetex        : true
plt.rc('text', usetex=True)
```

```python
d000 = np.loadtxt('Data00_LocalEncoded_20_10_2000.txt', delimiter='\t', unpack = True)
d110 = np.loadtxt('Data110_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
d111 = np.loadtxt('Data111_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
d200 = np.loadtxt('Data20_LocalEncoded_20_10_1000.txt', delimiter='\t', unpack = True)
dun0 = np.loadtxt('DataUn_LocalEncoded_20_10_2000.txt', delimiter='\t', unpack = True)
```

```python
t = d000[0][:-1]
d00 = d000[1][:-1]
d11 = [(d110[4][i] + d111[4][i])/2 for i in range(len(t))]
d20 = d200[6][:-1]
dun = dun0[1][:-1]
```

```python
fig, ax = plt.subplots(figsize=(6,5.25))

ax.plot(t,dun,marker='D',markevery=10,color='red',linestyle='dashed',label=r'Unencoded')

ax.plot(t,d00,marker='^',markevery=10,color='blue',linestyle='dashed',label=r'Encoded')

ax.plot(t,d11,marker='^',markevery=10,color='black',linestyle='dashdot',label=r'\noindent Encoded w/QEC')

ax.plot(t,d20,marker='^',markevery=10,color='green',linestyle='solid',label=r'\noindent Encoded w/QEC \\and Extraction')


ax.set_xlim((0,1.0))
ax.set_ylim((-0.02,1.04))
ax.set_xlabel(r'$\gamma_{m}t$',fontsize=22,labelpad=5)
ax.set_ylabel(r'$\mathcal{F}$',fontsize=22,labelpad=20)
plt.ticklabel_format(axis='y',style='sci',scilimits=(0,0))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig("Fig2d.pdf",bbox_inches='tight',dpi=600)
plt.show()
```
